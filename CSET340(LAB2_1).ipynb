```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files

# Upload the image
uploaded = files.upload()

# Get the uploaded file name
image_path = list(uploaded.keys())[0]

# Read the image in grayscale mode
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Check if the image is loaded correctly
if image is None:
    print("Error: Unable to load image.")
else:
    plt.imshow(image, cmap='gray')
    plt.title("Original Image")
    plt.axis('off')
    plt.show()
```

```python
# Resize the image using different interpolation methods
resized_linear = cv2.resize(image, (200, 200), interpolation=cv2.INTER_LINEAR)
resized_nearest = cv2.resize(image, (200, 200), interpolation=cv2.INTER_NEAREST)
resized_cubic = cv2.resize(image, (200, 200), interpolation=cv2.INTER_CUBIC)

# Display resized images
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(resized_linear, cmap='gray')
plt.title("Linear Interpolation")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(resized_nearest, cmap='gray')
plt.title("Nearest Neighbors")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(resized_cubic, cmap='gray')
plt.title("Cubic (Polynomial) Interpolation")
plt.axis('off')

plt.show()
```

```python
# Apply different types of blurring
box_blur = cv2.blur(image, (5, 5))
gaussian_blur = cv2.GaussianBlur(image, (5, 5), 0)
median_blur = cv2.medianBlur(image, 5)

# Display the blurred images
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(box_blur, cmap='gray')
plt.title("Box Blurring")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(gaussian_blur, cmap='gray')
plt.title("Gaussian Blurring")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(median_blur, cmap='gray')
plt.title("Median (Adaptive) Blurring")
plt.axis('off')

plt.show()
```
